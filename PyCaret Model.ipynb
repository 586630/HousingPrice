```python
# lists the files in the folder
import os
print(os.listdir("./data2"))
```

```python
train = pd.read_csv("./data2/train.csv")
test = pd.read_csv("./data2/test.csv")
```

```python
sp = pd.DataFrame({'SalePrice': range(0, test.shape[0] - 1)})
```

```python
test = pd.concat([test, sp], axis=1)
```

# PyCaret

```python
from pycaret.classification import *
```

```python
train.describe()
```

```python
train = train.drop(columns="Id", axis=1)
```

```python
data = setup(data=train, test_data=test, target='SalePrice', train_size=0.99, fold_shuffle=True, fold = 2, normalize=True)
```

```python
best_models = compare_models(exclude = ['svm', 'gbc', 'knn', 'dt', 'ridge'], n_select=4)
```

```python
best_models
```

```python
tuned_best_model = tune_model(best_models[0])
```

```python
blender_specific = blend_models(estimator_list = best_models[1:3], method = 'soft')
```

```python
data = plot_model(blender_specific)
```

```python
plot_model(blender_specific, plot = 'confusion_matrix')
```

```python
prediction_holdout = predict_model(blender_specific)
```

```python
save_model(blender_specific,'best_model')
```
